```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('coffee_production.csv')
print(data.columns)
```

```python
data.head(10)
```

```python
data.isnull
```

```python
#clean data
data['year'] = pd.to_datetime(data['year'], format = '%Y')
data['production'] = pd.to_numeric(data['production'])
data['price'] = pd.to_numeric(data['price'])
```

```python
grouped_data = data.groupby(['year','country']).sum().reset_index()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = grouped_data,x='year',y='production',hue='country')
plt.title('Coffee Production Trends')
plt.xlabel('Year')
plt.ylabel('Production')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='year', y='price', hue='country')
plt.title('Coffee Price Trends')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```
